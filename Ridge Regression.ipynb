```python
import mglearn
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
X,y = mglearn.datasets.load_extended_boston()
```

```python
from sklearn.model_selection import train_test_split
X_train,X_test,y_train,y_test = train_test_split(X,y,random_state=0)
```

```python
from sklearn.linear_model import Ridge
rd = Ridge().fit(X_train,y_train)
```

```python
rd.score(X_train,y_train)
```

```python
rd.score(X_test,y_test)
```

```python
rd10 = Ridge(alpha=10).fit(X_train,y_train)
```

```python
print("Training Set Score: {}".format(rd10.score(X_train,y_train)))
```

```python
print("Test Set Score: {}".format(rd10.score(X_test,y_test)))
```

```python
plt.figure(figsize=(15,10))
plt.plot(rd.coef_,'^',label = 'Ridge with alpha = 1')
plt.plot(rd10.coef_,'v',label='Ridge with alpha = 10')
plt.xlabel("Coefficient Index")
plt.ylabel("Coefficient Magnitude")
plt.ylim(-10,20)
plt.legend()
```
